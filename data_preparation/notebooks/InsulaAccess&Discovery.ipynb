```python
import pandas as pd
import requests
import json
import re
import time

# SPRÁVNE definuj HEADERS ako DICT (nie set!)

BASE_URL = "https://phisat2.insula.earth"
collection_url = f"{BASE_URL}/secure/api/v2.0/collections/7"

def get_all_l1_files():
    all_files = []
    url = f"{BASE_URL}/secure/api/v2.0/platformFiles/search/parametricFind?collection={collection_url}&sort=filename"
    
    page = 0
    while url:
        print(f"Načítavam stránku {page}...")
        r = requests.get(url, headers=HEADERS, verify=True)
        data = r.json()
        all_files.extend(data["_embedded"]["platformFiles"])
        
        next_link = data["_links"].get("next", {})
        url = next_link.get("href") if next_link else None
        page += 1
        time.sleep(0.1)
    
    print(f"Nájdených L1 súborov: {len(all_files)}")
    return all_files

def extract_l0_id(image_id):
    match = re.search(r'PHISAT-2_L0_(\d+)_', image_id)
    return match.group(1) if match else None

def check_l1_access(csv_path, l1_files):
    df = pd.read_csv(csv_path)
    print(f"Načítaných L0 súborov: {len(df)}")
    
    results = []
    for idx, row in df.iterrows():
        image_id = row['image_id']
        l0_id = extract_l0_id(image_id)
        
        if not l0_id:
            results.append({
                'image_id': image_id, 'l0_id': None, 'l1_available': False, 
                'l1_filename': None, 'l1_id': None
            })
            continue
        
        l1_pattern = f"PHISAT-2_L1_{l0_id}_"
        l1_matches = [f for f in l1_files if l1_pattern in f["filename"]]
        
        if l1_matches:
            results.append({
                'image_id': image_id, 'l0_id': l0_id, 'l1_available': True,
                'l1_filename': l1_matches[0]["filename"], 'l1_id': l1_matches[0]["id"]
            })
        else:
            results.append({
                'image_id': image_id, 'l0_id': l0_id, 'l1_available': False,
                'l1_filename': None, 'l1_id': None
            })
    
    return pd.DataFrame(results)

# Spusti
l1_files = get_all_l1_files()
results = check_l1_access("/home/van6/bcs-thesis/dataset.csv", l1_files)

print(f"\n=== VÝSLEDKY ===")
print(f"Prístup k L1: {results['l1_available'].sum()}/{len(results)} ({results['l1_available'].mean():.1%})")

results.to_csv("/home/van6/bcs-thesis/l1_access_check.csv", index=False)
print("Uložené do: l1_access_check.csv")
```

### Download file

We're getting a valid `file_id` from the request done in the last section. A custom one can be placed here if needed

```python
file_id=run_request_dict['_embedded']['platformFiles']
print(f"file_id: {file_id}")
```

In this section the file download is done through the help of a function to which we have to provide the following parameters:
- `url`: the path to a specific file inside the platform, this contains the `file_id`.
- `local_filename`: the name of the downloaded file
- `customHeader`: through this we can pass the `HEADERS` containing our auth, as the request is done inside the function.

```python
def download_file(url, local_filename, customHeader, checkCert=True):
    print(f'{url} started')
    with requests.get(url,headers=customHeader, stream=True, verify=checkCert ) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):  
                f.write(chunk)

    print(f'download {url} done')
    return local_filename
```

```python
file_to_download = BASE_URL + "/secure/api/v2.0/platformFiles/" + str(file_id)+"/dl"
download_file(file_to_download,'downloadTestFile.tif',HEADERS)
```
